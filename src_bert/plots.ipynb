## Plotting

```python
import pandas as pd
from helpers import plot_results
# BERT model
result_train = "../results/distilbert_v4_train.csv"
result_tune = "../results/distilbert_v4_tune.csv"
# load csv data to dict of list
data_train = pd.read_csv(result_train)
data_tune = pd.read_csv(result_tune)
train_maxiter = data_train["iter"].max()
data_tune["iter"] = data_tune["iter"].apply(lambda x:x+train_maxiter)
data = data_train.append(data_tune, ignore_index=True)
ax = plot_results(data)
```
